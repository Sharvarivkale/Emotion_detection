```python
from keras.utils import to_categorical
from keras.preprocessing.image import load_img, img_to_array
from keras.models import Sequential
from keras.layers import Dense,Conv2D,Dropout,Flatten,MaxPooling2D
import os
import pandas as pd
import numpy as np
```

```python
Train_dir="images/train"
test_dir="images/validation"
```

```python
def createdf(dir):
    image_path=[]
    labels=[]

    for label in os.listdir(dir):
        for imagename in os.listdir(os.path.join(dir,label)):
            image_path.append(os.path.join(dir,label,imagename))
            labels.append(label)
        print(label,"completed")
    return image_path,labels
```

```python
train=pd.DataFrame()
train['image_path'],train['label']=createdf(Train_dir)
```

```python
print(train)
```

```python
test=pd.DataFrame()
test['image_path'],test['label']=createdf(test_dir)
```

```python
print(test)
```

```python
from tqdm.notebook import tqdm
```

```python
def extract_feactures(images):
    feactures=[]
    for image in tqdm(images):
        img=load_img(image, color_mode='grayscale')
        img=np.array(img)
        feactures.append(img)
    feactures=np.array(feactures)
    feactures=feactures.reshape(len(feactures),48,48,1)
    return feactures
```

```python
train_images_path=extract_feactures(train["image_path"] )
```

```python
test_images_path=extract_feactures(test["image_path"] )
```

```python
x_train=train_images_path/255.0
x_test=test_images_path/255.0
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
le.fit(train['label'])
```

```python
y_train=le.transform(train['label'])
y_test=le.transform(test['label'])
```

```python
y_train=to_categorical(y_train,num_classes=7)
y_test=to_categorical(y_test,num_classes=7)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense

input_shape = (48, 48, 1)

model = Sequential()

model.add(Conv2D(128, (3,3), activation='relu', input_shape=input_shape))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.4))

model.add(Conv2D(256, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.4))

model.add(Conv2D(512, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.4))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.4))

# 🔥 FIXED OUTPUT LAYER
model.add(Dense(7, activation='softmax'))

# 🔥 FIXED LOSS FUNCTION
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

model.fit(
    x_train,
    y_train,
    batch_size=128,
    epochs=3,
    validation_data=(x_test, y_test)
)
```

```python
model_json = model.to_json()

with open("emotiondetector.json", "w") as json_file:
    json_file.write(model_json)

model.save("emotiondetector.h5")
```
